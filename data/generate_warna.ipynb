```python
import csv

# Baca file CSV yang sudah dibuat sebelumnya
nama_file_input = "data_warna.csv"
data_warna = []

with open(nama_file_input, mode='r', newline='', encoding='utf-8') as file:
    reader = csv.reader(file)
    header = next(reader)  # Baca header
    for row in reader:
        data_warna.append(row)

# Tentukan nama file CSV untuk hasil kombinasi
nama_file_output = "kombinasi_warna.csv"

# Buat set kosong untuk menyimpan kombinasi warna yang sudah ada
kombinasi_sudah_ada = set()

# Buat file CSV untuk menulis kombinasi warna
with open(nama_file_output, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Kombinasi Warna"] + header[1:])  # Tulis header
    
    # Tulis setiap kombinasi warna unik ke dalam file CSV
    for i in range(len(data_warna)):
        for j in range(i + 1, len(data_warna)):
            # Dapatkan nama warna dari kedua baris data
            warna_1 = data_warna[i][0]
            warna_2 = data_warna[j][0]
            # Jika kombinasi ini belum ada dan kedua warnanya berbeda
            if (warna_1 + " " + warna_2) not in kombinasi_sudah_ada and (warna_2 + " " + warna_1) not in kombinasi_sudah_ada:
                kombinasi_sudah_ada.add(warna_1 + " " + warna_2)  # Tambahkan kombinasi ke set
                kombinasi_sudah_ada.add(warna_2 + " " + warna_1)  # Tambahkan kombinasi ke set
                
                # Gabungkan nilai-nilai dari kedua baris data
                kombinasi_baru = [warna_1 + " " + warna_2] + [", ".join(set(data_warna[i][k].split(", ") + data_warna[j][k].split(", "))) for k in range(1, len(data_warna[i]))]
                writer.writerow(kombinasi_baru)  # Tulis kombinasi ke file CSV

print("File CSV kombinasi berhasil dibuat:", nama_file_output)
```
